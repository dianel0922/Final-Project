```python
import gensim
from gensim.models import Word2Vec
from nltk.tokenize import sent_tokenize, word_tokenize
from nltk.tokenize.toktok import ToktokTokenizer
import pandas as pd
```

```python
# 這邊是未處理完的dataset，僅作演示使用
from sklearn.utils import shuffle

dataset = pd.read_csv('/Users/kiyas/Desktop/AI_Intro/Final_Project/dataset.csv')
dataset.drop(columns=['Unnamed: 0'], axis = 1, inplace = True)

train_neg = dataset[dataset['label']==0].iloc[0:round(34790*0.7)]
test_neg = dataset[dataset['label']==0].iloc[round(34790*0.7):]
train_pos = dataset[dataset['label']==1].iloc[0:round(28885*0.7)]
test_pos = dataset[dataset['label']==1].iloc[round(28885*0.7):]
train = pd.concat([train_neg, train_pos], axis = 0)
test = pd.concat([test_neg, test_pos], axis=0)
train = shuffle(train)
test = shuffle(test)
```

```python
train
```

```python
tokenizer = ToktokTokenizer()
corpus = [tokenizer.tokenize(document) for document in train['title']]
model = gensim.models.Word2Vec(corpus, min_count = 1,
                                            window = 5, sg = 1) # sg -> skim gram = true
model.train(corpus, total_examples=len(corpus), epochs = 50)
# model.save("word2vec.model")
```

```python
vector = model.wv
# vector.save("word2vec.wordvectors")
```
