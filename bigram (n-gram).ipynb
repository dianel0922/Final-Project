```python
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
```

```python
train = pd.read_csv('train.csv', index_col=0)
train
```

```python
train.groupby(['label']).size()
```

```python
test = pd.read_csv('test.csv', index_col=0)
test
```

```python
test.groupby(['label']).size()
```

```python
x_train = train.drop('label', axis=1)
y_train = train['label']
x_test = test.drop('label', axis=1)
y_test = test['label']
```

```python
# Bigram: TF-IDF
tfidf = TfidfVectorizer(ngram_range=(2, 2))

x_train_title_tf = tfidf.fit_transform(x_train['title'])
x_test_title_tf = tfidf.transform(x_test['title'])
```

```python
x_train_title_tf
```

```python
df_tfidf = pd.DataFrame(x_train_title_tf.todense(), columns = tfidf.get_feature_names()) 
df_tfidf.T
```

```python
# Bigram: CV
cv = CountVectorizer(ngram_range=(2, 2))
x_train_title_cv = cv.fit_transform(x_train['title'])
```

```python
x_train_title_cv
```

```python
df_cv = pd.DataFrame(x_train_title_cv.todense(), columns = cv.get_feature_names()) 
df_cv.T
```
